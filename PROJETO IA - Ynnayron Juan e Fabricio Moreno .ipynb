## ÁRVORE DE DECISÃO REGRESSÃO

```python
from sklearn.tree import DecisionTreeRegressor, export_graphviz
import graphviz
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error


X_treino, X_teste, y_treino, y_teste = train_test_split(x_admissao, y_admissao, test_size=0.3, random_state=42)

modelo_arvore = DecisionTreeRegressor()
```

```python
modelo_arvore.fit(X_treino, y_treino)
```

```python
pred_arvore = modelo_arvore.predict(x_admissao) # predict da arvore de decisão
pred_arvore
```

## GRÁFICO ÁRVORE

```python
dot_data = export_graphviz(modelo_arvore, max_depth = 3,  #Plot da árvore de decisão para melhor visualização da mesma
                      out_file=None,
                      feature_names=admissaoup.columns,
                      class_names=['Yes','No'],
                      filled=True, rounded=True,
                      special_characters=True)
graph = graphviz.Source(dot_data)
graph
```

## REGRESSÃO LINEAR

```python
from sklearn.linear_model import LinearRegression
Regressao_linear= LinearRegression()
Regressao_linear.fit(X_treino, y_treino)
```

```python
pred_reg_linear_X= Regressao_linear.predict(x_admissao)
pred_reg_linear_X
```

# GRÁFICO DE REGRESSÃO LINEAR

```python
# Cálculo da regressão linear para obter a linha 'y'
m, b = np.polyfit(pred_reg_linear_X, y_admissao, 1)
linha_y = m * pred_reg_linear_X + b  # Valores previstos pela linha

# Criação do gráfico
plt.figure(figsize=(8, 6))

#Plotagem de dispersão dos pontos pred_reg_linear_X no eixo x e y_admissao no eixo y
plt.scatter(pred_reg_linear_X, y_admissao, label='Pontos x', color='blue')

# Plotagem da linha 'y'
plt.plot(pred_reg_linear_X, linha_y, label='Linha y', color='red')

plt.xlabel('Valores de x (pred_reg_linear_X)')
plt.ylabel('Valores de y (y_admissao)')
plt.legend()
plt.title('Relação entre x e y')

plt.show()
```

# REDE NEURAL ARTIFICIAL


```python
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint as sp_randint
MODELO_RNA = MLPRegressor(activation= 'tanh', hidden_layer_sizes=(200, 100), learning_rate='adaptive', solver = 'adam')
MODELO_RNA.fit(X_treino, y_treino)
```

# DEFINIÇÃO DE HIPERPARAMETROS

A seguir foi utilizado da biblioteca RandomizedSearchCV onde a mesma realiza um busca aleatoria baseda no x e y de teste para os melhores hiperparametros para o modelo.


```python
param_dist = {
    'hidden_layer_sizes': [(50, 50), (100, 50, 25), (300, 200, 100)],
    'activation': ['relu', 'tanh'],
    'solver': ['adam', 'sgd'],
    'learning_rate': ['constant', 'adaptive'],
    'alpha': sp_randint(1, 10),  # Exemplo de distribuição aleatória para o parâmetro alpha
}

# Inicialização da pesquisa aleatória com validação cruzada
random_search = RandomizedSearchCV(MODELO_RNA, param_distributions=param_dist, n_iter=10, cv=3, random_state=42)

# Treinamento da pesquisa aleatória nos dados de treino
random_search.fit(X_treino, y_treino)

print("Melhores Parâmetros:", random_search.best_params_)
```

```python
RNA_PRED=MODELO_RNA.predict(x_admissao)
RNA_PRED
```

# ANÁLISE DE ACURÁCIA DOS MODELOS

```python
arvore_pontuação = (modelo_arvore.score(X_teste, y_teste))*100
arvore_pontuação
```

```python
reg_linear_pontuação = (Regressao_linear.score(X_teste, y_teste))*100
reg_linear_pontuação
```

```python
MODELO_RNA_pontuação = (MODELO_RNA.score(X_teste, y_teste))*100
MODELO_RNA_pontuação
```

```python
Modelos = ['Regressão Linear', 'Árvore de Decisão', 'RNA']
pontos = np.array([reg_linear_pontuação, arvore_pontuação, MODELO_RNA_pontuação])
plt.figure(figsize=(8, 6))
sns.barplot(x=Modelos, y=pontos)
plt.title('Acurácia dos Modelos')
plt.ylabel('Acurácia')
plt.ylim(0, 100)

plt.show()
```
